# Data visualization

```python
%matplotlib inline
import matplotlib.pyplot as plt

from torch.utils.data import DataLoader
from torchvision import transforms
```

Load a custom subclass of torchvision.datasets.MNIST that instead downloads the FashionMNIST dataset 

(waiting for [this commit](https://github.com/pytorch/vision/commit/eec5ba4405c8815bd1797619d9cc9276f81b76f4) be available in the pytorch version installable via pip).

```python
import os
import sys
sys.path.append(os.path.join(os.getcwd(), os.pardir, 'src'))
from data.fashion import FashionMNIST
```

Load some images and visualize them

```python
batch_size = 64

dataset = FashionMNIST('../data', train=True, download=True, transform=transforms.ToTensor())
loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
```

```python
fig, axes = plt.subplots(8, 8, figsize=(16, 16))

for ax, data, label in zip(axes.ravel(), *next(loader.__iter__())):
    ax.imshow(data.numpy().squeeze(), cmap='gray')
    ax.set_title(FashionMNIST.labels[label])
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
```
